## Paso 2: Limpieza de datos y tratamiento de valores faltantes

En esta sección abordaremos la columna `duration` (duración de la película):

- Primero, contamos cuántos valores faltantes (`NaN`) hay en esta columna.
- Luego, reemplazamos esos valores faltantes por la mediana de la duración, para no afectar el análisis con valores extremos.
- Finalmente, calculamos y mostramos la duración promedio de las películas, redondeada a dos decimales.

```python
# Contar valores faltantes (NaN) en la columna 'duration'
missing_duration = df['duration'].isna().sum()
print(f'Valores faltantes en duration: {missing_duration}')
```

```python
# Calcular la mediana de 'duration'
median_duration = df['duration'].median()

print(f'Mediana de duración: {median_duration}')

# Reemplazar los NaN por la mediana
df['duration'] = df['duration'].fillna(median_duration)
```

```python
# Calcular el promedio de 'duration'
mean_duration = round(df['duration'].mean(), 2)
print(f'Duración promedio de las películas: {mean_duration}')
```

### Interpretación de resultados

- Se encontraron **15** valores faltantes en la columna `duration`, lo que indica que algunas películas no tenían registrada su duración.
- Estos valores faltantes fueron reemplazados por la **mediana (103.0 minutos)** de la duración para evitar distorsionar el análisis con valores extremos.
- La duración promedio calculada después de la imputación es de **107.19 minutos**, lo que nos da una idea del tiempo típico de las películas en el dataset.

Este tratamiento garantiza que el análisis posterior de la duración sea más robusto y representativo.

## Paso 3: Análisis específico

En este paso realizaremos análisis más concretos sobre el dataset para responder preguntas específicas:

- ¿Cuántas películas fueron lanzadas en 2008 y tienen una duración entre 90 y 120 minutos (inclusive)?
- ¿Cuál es la mediana del presupuesto de todas las películas?

Estos análisis nos ayudarán a entender mejor algunas características puntuales de los datos.


```python
# Filtrar las películas lanzadas en 2008 con duración entre 90 y 120 minutos
peliculas_2008 = df[(df['title_year'] == 2008) & (df['duration'] >= 90) & (df['duration'] <= 120)]

# Contar cuántas cumplen esa condición
cantidad_peliculas_2008 = len(peliculas_2008)
print(f'Cantidad de películas en 2008 con duración entre 90 y 120 minutos: {cantidad_peliculas_2008}')
```

```python
# Crear una copia temporal de la columna como string
budget_temp = df['budget'].astype(str)

# Quitar el signo $ y comas solo en esta copia temporal
budget_temp = budget_temp.str.replace('$', '', regex=False).str.replace(',', '', regex=False)

# Convertir esa copia a numérico para calcular mediana
budget_temp_num = pd.to_numeric(budget_temp, errors='coerce')

# Calcular la mediana sin modificar df
median_budget = budget_temp_num.median()

print(f'Mediana presupuesto (sin modificar df): {median_budget:.0f}')
```

### Interpretación de resultados

- En el año **2008**, se encontraron **160** películas cuya duración está entre **90 y 120 minutos**. Esto indica la cantidad de producciones con duración estándar en ese año.
- La mediana del presupuesto de todas las películas del dataset es aproximadamente **$15000000**. La mediana es un buen indicador del presupuesto típico, ya que no se ve afectada por valores extremos o errores en los datos.

Estos análisis nos permiten entender mejor la distribución temporal y económica del dataset de películas.

## Paso 4: Retos

En este paso haremos uno o varios análisis adicionales, por ejemplo:

- ¿Qué relación hay entre el presupuesto y la duración?

- ¿Cuáles son los directores que más películas tienen en el dataset?

- ¿Hay películas con duración atípica (muy cortas o muy largas)?

Estas ideas nos ayudarán a profundizar y encontrar patrones interesantes en los datos.

```python
# Relación entre presupuesto y duración

import matplotlib.pyplot as plt
import matplotlib.ticker as mtick

# Limpiar y convertir presupuesto a numérico
budget_temp = df['budget'].astype(str)
budget_temp = budget_temp.str.replace('$', '', regex=False).str.replace(',', '', regex=False)
budget_temp_num = pd.to_numeric(budget_temp, errors='coerce')

# Filtrar presupuestos razonables (ej: hasta 400 millones USD)
filtered_df = df.copy()
filtered_df['budget_clean'] = budget_temp_num
filtered_df = filtered_df[(filtered_df['budget_clean'] <= 400_000_000) & (filtered_df['budget_clean'] > 0)]

# Crear gráfico de dispersión
plt.figure(figsize=(10, 6))
plt.scatter(filtered_df['budget_clean'], filtered_df['duration'], alpha=0.5)

plt.title('Presupuesto vs Duración')
plt.xlabel('Presupuesto (en millones de USD)')
plt.ylabel('Duración (minutos)')

# Formatear eje X para mostrar en millones
plt.gca().xaxis.set_major_formatter(mtick.FuncFormatter(lambda x, _: f'{x * 1e-6:.0f}M'))

plt.grid(True)
plt.tight_layout()
plt.show()
```

## Análisis

- No hay relación clara entre presupuesto y duración.

- La mayoría de películas duran entre 90 y 140 minutos.

- Películas con presupuesto alto también suelen tener duraciones estándar.

- Existen películas de bajo presupuesto con duraciones largas.

- Se observan algunos valores atípicos con duraciones extremas.

En general no existe una relación directa entre el presupuesto y la duración de una película. La duración parece responder más a decisiones creativas o de formato que al tamaño del presupuesto.

## Análisis

Podemos evidenciar que la persona con mayor cantidad de películas dirigidas es Steven Spielberg seguido de Woody Allen, Martin Scorsese y Clint Eastwood.

```python
# Películas con duraciones atípicas

# Calcular cuartiles y rango intercuartílico (IQR) para la duración en minutos
q1 = df['duration'].quantile(0.25)
q3 = df['duration'].quantile(0.75)
iqr = q3 - q1

# Definir límites inferior y superior para detectar outliers
limite_inferior = q1 - 1.5 * iqr
limite_superior = q3 + 1.5 * iqr

# Filtrar películas con duración atípica (fuera de los límites)
outliers_duracion = df[(df['duration'] < limite_inferior) | (df['duration'] > limite_superior)]

# Mostrar cantidad de películas atípicas y sus títulos con duración en minutos
print(f'Películas con duración atípica (en minutos): {len(outliers_duracion)}')
print(outliers_duracion[['movie_title', 'duration']].sort_values(by='duration'))
```

## Análisis de películas con duraciones extremas

Al revisar las películas con duraciones atípicas, encontramos lo siguiente:

- Las películas más cortas duran menos de 20 minutos, por ejemplo:
  - *Shaun the Sheep* y *The Touch* duran 7 minutos cada una.
  - *Robot Chicken* tiene una duración de 11 minutos.
  - Estos títulos probablemente corresponden a cortometrajes o episodios especiales.

- Por otro lado, las películas más largas superan ampliamente los 300 minutos, con títulos como:
  - *The Legend of Suriyothai* (300 minutos).
  - *Heaven's Gate* (325 minutos).
  - *Blood In, Blood Out* (330 minutos).
  - *Carlos* (334 minutos).
  - *Trapped* es el caso más extremo con 511 minutos (más de 8 horas), lo que podría tratarse de una miniserie o edición extendida.

Estas duraciones extremas representan valores atípicos respecto a la mayoría, que suelen durar entre 90 y 140 minutos. Identificarlas es útil para entender la diversidad del dataset y decidir si se incluyen o excluyen en análisis posteriores.
